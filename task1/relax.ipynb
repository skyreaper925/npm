```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Создаём списки для хранения данных
T_total = []
T_cold = []
T_hot = []
n_cold = []
n_hot = []
S_total = []
S_cold = []
S_hot = []
```

```python
# Читаем файл
with open('conf.txt', 'r') as file:
    for line in file:
        # Разбиваем строку на элементы и преобразуем в float
        values = [float(x) for x in line.split()]
        T_total.append(values[0])
        T_cold.append(values[1])
        T_hot.append(values[2])
        n_cold.append(values[3])
        n_hot.append(values[4])
        S_total.append(values[5])
        S_cold.append(values[6])
        S_hot.append(values[7])
```

```python
# Создаем фигуру с тремя подграфиками
plt.figure(figsize=(15, 10))
t = np.arange(1, len(T_total)+1)

# Первый график: T_total, T_cold, T_hot
plt.scatter(t, T_total, label='T_total', color="violet")
plt.scatter(t, T_cold, label='T_cold', color="blue")
plt.scatter(t, T_hot, label='T_hot', color="red")
plt.xlabel(r'Время, 0.05 $\tau$')
plt.ylabel('Температура')
plt.legend()
plt.grid(True)

# Показываем графики
plt.show()
```

```python
plt.figure(figsize=(15, 10))
# Второй график: n_cold, n_hot
plt.scatter(t, n_cold, label='n_cold', color="blue")
plt.scatter(t, n_hot, label='n_hot', color="red")
plt.xlabel('Время')
plt.ylabel('Концентрация')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(t, S_total, label='S_total', color="violet")
plt.scatter(t, S_cold, label='S_cold', color="blue")
plt.scatter(t, S_hot, label='S_hot', color="red")
plt.xlabel(r'Время, 0.05 $\tau$')
plt.ylabel('Энтропия')
plt.legend()
plt.grid(True)

plt.show()
```
